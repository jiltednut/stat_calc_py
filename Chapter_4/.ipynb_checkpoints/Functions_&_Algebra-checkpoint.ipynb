```python
# Exercise 4.01: Function identification from plots
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Horizontal line
"""
This a function because for any value of x, a vertical line will
only intersect with one value of y

The domain is the set of all Real numbers
The range is the set of {2}

The formula for the function is f(x) = 2

The formula has no root as the range does not contain the value 0

"""


x = np.linspace(-5, 5, 1000)
y = np.linspace(2, 2, 1000)

plt.plot(x, y)
plt.ylim([-2,5])
plt.xlim([-6, 6])
plt.show()
```

```python
# 2. Rotated quadratic curve
"""
The graph is not a valid function because of any positive input value
there are 2 output values.

"""

```

```python
# 3. Straight Line
"""
The graph is of a valid function as it passes the vertical line test
The domain is the set of all real numbers
The range is the set of all real numbers
The functions is: f(x) = x

The formula for the root for is f(0) = 0

"""

plt.plot(x, x)
plt.show()
```

```python
# 4. Quadratic curve
"""
The plot is of a valid function as it passes the vertical line test.
The domain is the set of all real numbers.
The range is the set of all positive real numbers.
The function is: f(x) = x^2

The formula for the root is: f(0) = 0

"""

plt.plot(x, (x**2))
plt.show()
```

```python
# Function Transformations
plt.plot(x, x**2, c = 'b')
plt.plot(x, (x-2)**2, c = 'r')
plt.xlim([-5, 5])
plt.ylim([0, 5])
plt.show()
```

```python
# Scaling
plt.plot(x, x**2, c = 'b')
plt.plot(x, -((x/2)**2), c = 'r')
plt.xlim([-5, 5])
plt.ylim([-5, 5])
plt.show()
```

```python
# Exercise 4.02: Function Transformation Identification
# 1. First transformation of a cubic curve
"""
The curves has undergone a shifting transformation: moving by 2 on the x axis

"""
plt.plot(x, x**3 - x, c = 'b')
plt.plot(x, (x + 2)**3 - (x+2), c = 'r')
plt.xlim([-5, 5])
plt.ylim([-5, 5])
plt.show()
```

```python
# 2. Second transformation of a cubic curve
"""
The curves has undergone a shifting transformation (composite shift): 
    moving by 2 on the x axis and 2 on the Y axis

"""
plt.plot(x, x**3 - x, c = 'b')
plt.plot(x, ((x - 2)**3 - (x-2))+2, c = 'r')
plt.xlim([-5, 5])
plt.ylim([-5, 5])
plt.show()
```

```python
# First transformation of a sine wave
plt.plot(x, np.sin(x), c = 'b')
plt.plot(x, np.sin(x)*2, c = 'r')
plt.xlim([-5, 5])
plt.ylim([-5, 5])
plt.show()
```

```python
# Second transformation of a sine wave
plt.plot(x, np.sin(x), c = 'b')
plt.plot(x, np.sin(x*2), c = 'r')
plt.xlim([-5, 5])
plt.ylim([-5, 5])
plt.show()
```

```python
# Systems of Equations
def solve_eq_sys(coeff_matrix, c):
    try:
        inv_matrix = np.linalg.inv(coeff_matrix)
        return inv_matrix.dot(c)
    except np.linalg.LinAlgError:
        return False    
    
coeff_matrix = np.array([[1, 3, -2],
                        [3, 5, 6],
                        [2, 4, 3]])
c = np.array([1, 31, 19])

solve_eq_sys(coeff_matrix, c)
```

```python
coeff_matrix = np.array([[1, 3, -2],
                         [3, 5, 6],
                         [2, 6, -4]])

c = np.array([1, 31, 19])

solve_eq_sys(coeff_matrix, c)
```

```python
# Activity 4.01: Multi-Variate Break Even Analysis
x = Symbol('x') # num_burgers
y = Symbol('y') # burger_sale_price
burger_cost_price = 6.56
overhead = 1312.13
revenue = x * y
town_budget = 4000

profit = (x * (y - burger_cost_price)) - overhead
# (x * (y - burger_cost_price)) = overhead
```

```python
# x = 4000 / y  -  the number of burgers = revenue / sale_price
```

```python
solve([x * (y - burger_cost_price) - overhead, x * y - 4000], x, y)
```

```python
# 4.
def solve_function(num_burgers, burger_sale_price):
    demand = town_budget / burger_sale_price
    if num_burgers > demand:
        return town_budget - (num_burgers * burger_cost_price) - overhead
    
    return (num_burgers * (burger_sale_price - burger_cost_price)) - overhead
```

```python
solve_function(410, 9.77)
```

```python
# 5.
x = np.arange(300, 500, 2)
burger_sale_price = 9.76
profits_976 = [solve_function(x, burger_sale_price) for x in x]
profits_976

plt.plot(x, profits_976)
plt.axhline(0, c = 'r')
plt.show()
```

```python
# 6. 
profits_999 = [solve_function(x, 9.99) for x in x]
profits_999

plt.plot(x, profits_976, c = 'k')
plt.plot(x, profits_999)
plt.axhline(0, c = 'r')
plt.show()

# A price of 9.99 would breakeven approx 30 burgers sooner.
```
