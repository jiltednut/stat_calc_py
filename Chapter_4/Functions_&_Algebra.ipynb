```python
# Exercise 4.01: Function identification from plots
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Horizontal line
"""
This a function because for any value of x, a vertical line will
only intersect with one value of y

The domain is the set of all Real numbers
The range is the set of {2}

The formula for the function is f(x) = 2

The formula has no root as the range does not contain the value 0

"""


x = np.linspace(-5, 5, 1000)
y = np.linspace(2, 2, 1000)

plt.plot(x, y)
plt.ylim([-2,5])
plt.xlim([-6, 6])
plt.show()
```

```python
# 3. Straight Line
"""
The graph is of a valid function as it passes the vertical line test
The domain is the set of all real numbers
The range is the set of all real numbers
The functions is: f(x) = x

The formula for the root for is f(0) = 0

"""

plt.plot(x, x)
plt.show()
```

```python
# 4. Quadratic curve
"""
The plot is of a valid function as it passes the vertical line test.
The domain is the set of all real numbers.
The range is the set of all positive real numbers.
The function is: f(x) = x^2

The formula for the root is: f(0) = 0

"""

plt.plot(x, (x**2))
plt.show()
```

```python
# Function Transformations
plt.plot(x, x**2, c = 'b')
plt.plot(x, (x-2)**2, c = 'r')
plt.xlim([-5, 5])
plt.ylim([0, 5])
plt.show()
```

```python
# Scaling
plt.plot(x, x**2, c = 'b')
plt.plot(x, -((x/2)**2), c = 'r')
plt.xlim([-5, 5])
plt.ylim([-5, 5])
plt.show()
```

```python
# Exercise 4.02: Function Transformation Identification
# 1. First transformation of a cubic curve
"""
The curves has undergone a shifting transformation: moving by 2 on the x axis

"""
plt.plot(x, x**3 - x, c = 'b')
plt.plot(x, (x + 2)**3 - (x+2), c = 'r')
plt.xlim([-5, 5])
plt.ylim([-5, 5])
plt.show()
```

```python
# 2. Second transformation of a cubic curve
"""
The curves has undergone a shifting transformation (composite shift): 
    moving by 2 on the x axis and 2 on the Y axis

"""
plt.plot(x, x**3 - x, c = 'b')
plt.plot(x, ((x - 2)**3 - (x-2))+2, c = 'r')
plt.xlim([-5, 5])
plt.ylim([-5, 5])
plt.show()
```

```python
# First transformation of a sine wave
plt.plot(x, np.sin(x), c = 'b')
plt.plot(x, np.sin(x)*2, c = 'r')
plt.xlim([-5, 5])
plt.ylim([-5, 5])
plt.show()
```

```python
# Second transformation of a sine wave
plt.plot(x, np.sin(x), c = 'b')
plt.plot(x, np.sin(x*2), c = 'r')
plt.xlim([-5, 5])
plt.ylim([-5, 5])
plt.show()
```
