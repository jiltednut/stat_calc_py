```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Scatterplot
x = [1, 2, 3, 1.5, 2]
y = [-1, 5, 2, 3, 0]
plt.scatter(x, y)
plt.show()
```

```python
# Scatterplot customisations
sizes = [10, 40, 60, 80, 100]
colors = ['r', 'b', 'y', 'g', 'k']

plt.scatter(x, y, s = sizes, c = colors)
plt.show()
```

```python
# Line graphs
x = np.linspace(0, 10, 1000)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

```python
# Plotting multiple line graphs in one plot

x = np.linspace(1, 10, 1000)

linear_line = x
log_curve = np.log(x)
sin_wave = np.sin(x)

curves = [linear_line, log_curve, sin_wave]
colors = ['k', 'r', 'b']
styles = ['-', '--', ':']

for curve, color, style in zip(curves, colors, styles):
    plt.plot(x, curve, c = color, linestyle = style)
plt.show()
```

```python
# Bar charts
labels = ['Type 1', 'Type 2', 'Type 3']
counts = [2, 3, 5]

plt.bar(labels, counts)
plt.show()
```

```python
# Stacked & Grouped bar charts
type_1 = [1, 1]
type_2 = [1, 2]
type_3 = [2, 3]

counts = [type_1, type_2, type_3]
```

```python
locations = np.array([0, 1, 2])
width = 0.3

bars_a = plt.bar(locations - (width / 2), 
                 [my_type[0] for my_type in counts], width = width)
bars_b = plt.bar(locations + (width / 2), 
                 [my_type[1] for my_type in counts], width = width)

plt.xticks(locations, ['Type 1', 'Type 2', 'Type 3'])
plt.legend([bars_a, bars_b], ['Type A', 'Type B'])
plt.show()
```

```python
bars_a = plt.bar(locations, [my_type[0] for my_type in counts])
bars_b = plt.bar(locations, [my_type[1] for my_type in counts],
               bottom =[my_type[0] for my_type in counts])
plt.xticks(locations, ['Type 1', 'Type 2', 'Type 3'])
#plt.legend([bars_a, bars_b], ['Type A', 'Type B'])
plt.show()
```

```python
# Exercise 2.04
import scipy.stats as stats
```

```python
samples = np.random.normal(0, 1, size = 1000)
```

```python
x = np.linspace(samples.min(), samples.max(), 1000)
y = stats.norm.pdf(x)

plt.hist(samples, alpha = 0.2, bins = 20, density = True)
plt.plot(x, y)
plt.show()
```

```python
samples = np.random.beta(2, 5, size = 1000)
x = np.linspace(samples.min(), samples.max())

y = stats.beta.pdf(x, 2, 5)

plt.hist(samples, alpha = 0.2, bins = 20, density = True)
plt.plot(x, y)
plt.show()
```

```python
samples = np.random.gamma(1, size = 1000)
x = np.linspace(samples.min(), samples.max())

y = stats.gamma.pdf(x, 1)

plt.hist(samples, alpha = 0.2, bins = 20, density = True)
plt.plot(x, y)
plt.show()
```

```python
# Visualisaton with Seaborn
x = np.random.normal(0, 1, 1000)
y = np.random.normal(5, 2, 1000)

df = pd.DataFrame({'Column 1': x, 'Column 2': y})
df.head()
```

```python
import seaborn as sns

sns.jointplot(x = 'Column 1', y = 'Column 2', data = df)
plt.show()
```
